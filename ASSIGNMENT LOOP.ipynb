```python
ANS- In general, we should use a for loop when you know how many times the loop should run. 
If we want the loop to break based on a condition other than the number of times it runs, you should use a while loop. while loop always chek condition and if 
condition is true then while loop works.
```

```python
l = [1,2,3,4,5]
for i in l:
    print(i)
```

```python
l2 = ["abhay", "kumar" , "ranjan" , "pappu"]
print(l2)
```

```python
for i in l2:
    if i == "kumar":
        break
    print(i)
    
```

```python
# Sum of natural numbers

num = 10

if num < 0:
    print("Enter a positive number")
else:
    sum = 0
   # use while loop to iterate until zero
    while(num > 0):
        sum += num
        num -= 1
    print("The sum is", sum)
```
